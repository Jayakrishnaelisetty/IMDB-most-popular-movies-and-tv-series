### Missing Values handling methods.<a href="#Missing-Values-handling-methods." class="anchor-link">¶</a>

-   Delete feature / column:- If columns have more than 60% as null
    values then entire column can be dropped.
-   Numeric Type Feature:-
-   If feature/column does not contain outliers then replace missing
    values by mean.
-   If feature/column contain outliers then replace missing values by
    median. ( Median is not affected by outliers)
-   Categorical Type Feature:-
-   Replace missing values by Mode.( we can't compute Mean & Median of
    categorical features )
